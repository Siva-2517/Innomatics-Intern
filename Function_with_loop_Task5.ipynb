```python
#Problem 1: Social Media – Post Engagement Analyzer

def analyze_post(likes):
    total = 0
    for like in likes:
        total+= like

    print("Total Likes:", total)

    if total >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


likes = [200,300,150,400]
analyze_post(likes)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker

def analyze_rainfall(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", int(average))

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall = [60, 75, 80, 73] 
required_level = 70

analyze_rainfall(rainfall, required_level)
```

```python
#Problem 4: Social Media – Duplicate Account Detection

def detect(usernames):
    unique= set(usernames)

    if len(unique) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["user1","user2","user1","user3"]
detect(users)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium(prices):
    premium = [price for price in prices if price > 2000]
    print("Premium Crops:", premium)


crop_prices = [1500, 2500, 1800, 3200]
filter_premium(crop_prices)
```

```python
#Problem 7: System Monitoring – Application Health Checker

def check_system(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

check_system(7)
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker

def validate(amount):
    daily_limit = 50000
    print("Transaction Amount:", amount)

    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

validate(60000)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System

def eligibility(attendance):
    percentage = (sum(attendance) / len(attendance)) * 100

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:",
          "Eligible" if percentage >= 75 else "Not Eligible")


record = [1, 1, 0, 1, 1]
eligibility(record)
```

```python
#Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


units_consumed = 250
bill_amount, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", bill_amount)
print("Usage Level:", usage_status)
```
