```python
#Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
high_score = max(employees.values())

top_performers = []

for name, score in employees.items():
    if score == high_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      f"(Score: {high_score})")
```

```python
#Problem Statement 2: Search Query Keyword Analysis

import string

query = "Buy mobile phone buy phone online"

query = query.lower()

query = query.translate(str.maketrans('', '', string.punctuation))

words = query.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
#Problem Statement 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # check if even
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Problem Statement 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

print("Email Domain Usage:")
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#Problem Statement 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

average = sum(sales) / len(sales)

threshold = average * 1.29 

for i, value in enumerate(sales):
    if value >= threshold:
        print(f"Day {i + 1}: {value}")
```

```python
#Problem Statement 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

for uid in user_ids:
    if uid in id_count:
        id_count[uid] += 1
    else:
        id_count[uid] = 1

for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
