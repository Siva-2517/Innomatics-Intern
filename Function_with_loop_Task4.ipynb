```python
#Problem 1: Mobile Recharge Validation System

plans = [199, 299, 399, 599]

while True:
    try:
        a = int(input("Enter recharge amount: "))
        if a >= 50 and a in plans:
            print("Recharge Successful ")
            break
        print("Invalid! Must be ≥50 and one of", plans)
    except ValueError:
        print("Enter a valid number.")
```

```python
#Problem 2: Inventory Reorder Alert System

def check(products):
    for item, qty in products.items():
        if qty < 15:
            print(f"{item}: Reorder Alert")
        else:
            print(f"{item}: Stock OK")

stock = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check(stock)
```

```python
#Problem 3: Student Result Processing System

def check(marks):
    avg = sum(marks) / len(marks)
    return "Pass" if avg >= 50 else "Fail"

marks = [60, 45, 70, 55]
print(check(marks))
```

```python
#Problem 4: Cab Fare Estimator with Retry Option

def fare(km, peak):
    f = 50 + 12*km
    return f*1.25 if peak else f

while True:
    try:
        k = float(input("Distance km: "))
        p = input("Peak hour? (y/n): ").lower() == 'y'
        if k<0: print("Invalid!"); continue
        print(f"Total Fare: ₹{fare(k,p):.2f}"); break
    except ValueError: print("Enter number!")
```

```python
#Problem 5: Employee Attendance Eligibility 

def check(att):
    total = len(att)
    present = 0
    for a in att:
        if a.upper() == "P":
            present += 1
    percent = (present / total) * 100
    return "Eligible" if percent >= 75 else "Not Eligible"

attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
print(check(attendance))
```

```python
#Problem 6: Password Strength Checker

def check(p):
    if len(p)<8: return "Weak: too short"
    d = any(c.isdigit() for c in p)
    s = any(c in "@#$" for c in p)
    return "Strong" if d and s else "Weak: missing digit or special char"

pwd = input("Enter password: ")
print(check(pwd))
```
