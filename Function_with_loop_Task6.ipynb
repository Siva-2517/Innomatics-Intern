```python
#Problem Statement 1:Smart Parking Lot Management System

def manage_parking(capacity, logs):
    current_vehicles = 0
    peak_vehicles = 0

    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1
        peak_vehicles = max(peak_vehicles, current_vehicles)
        if current_vehicles > capacity:
            print("âš  ALERT: Parking Capacity Exceeded!")

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_vehicles)

    if current_vehicles < capacity:
        print("Parking Status: Available")
    else:
        print("Parking Status: Full")

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, vehicle_logs)
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator

def estimate(distance, traffic, weather):
    speed = 20  
    base_time = (distance / speed) * 60

    if traffic.lower() == "low":
        base_time += 5
    elif traffic.lower() == "medium":
        base_time += 15
    elif traffic.lower() == "high":
        base_time += 25

    if weather.lower() == "rainy":
        base_time += 6
    elif weather.lower() == "storm":
        base_time += 15

    print("Estimated Delivery Time:", int(base_time), "minutes")

distance = 8
traffic = "High"
weather = "Rainy"

estimate(distance, traffic, weather)
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

def analyze_booking(total_seats, booked_seats):
    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    print("\nOccupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    elif occupancy >= 40:
        print("Show Status: Filling Fast")
    else:
        print("Show Status: Available")

    if occupancy >= 90:
        print("Suggestion: Consider Opening Additional Show")



total_seats = int(input("Enter Total Seats: "))
booked_count = int(input("Enter Number of Booked Seats: "))

booked_seats = [1] * booked_count

analyze_booking(total_seats, booked_seats)
```

```python
# Problem Statement 4:Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0
    
    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    print("\nAverage CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor

def analyze_resource_usage(resource_usage):
    overused = []

    threshold = 8

    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

analyze_resource_usage(resource_usage)
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller

def manage(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


event_capacity = 100
registrations = 105

manage(event_capacity, registrations)
```
