```python
#1. User Login Check


#Stored credentials
username = "admin"
password = "1234"

# Taking input from user
input_name = input("Enter username: ")
input_password = input("Enter password: ")

# Checking login
if username == input_name and password == input_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = sum(1 for mark in marks if mark >= 50)
fail_count = len(marks) - pass_count

print(f"Total Students: {len(marks)}")
print(f"Pass Students: {pass_count}")
print(f"Fail Students: {fail_count}")

```

```python
#3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]

print("Cleaned Names:", cleaned_names)
```

```python
#4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    status = "Long Message" if length > 10 else "Short Message"
    print(f"{msg} → Length: {length} → {status}")
```

```python
#5. Error Message Detector


logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = sum(1 for log in logs if log=="ERROR")

print("Total ERROR entries:", error_count)
```
