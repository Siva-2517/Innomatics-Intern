```python
#Problem 1: Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def disp(self):
        print(f"{self.name} is {self.status}")

light= input("Light Name: ")
action = input("Action (ON/OFF): ").upper()

light = SmartLight(light)

if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid Action!")

light.disp()
```

```python
#Problem 2: Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, depart):
        self.name = name
        self.emp_id = emp_id
        self.depart = depart

    def disp_id(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.depart}")


name = input("Employee Name: ")
id = input("Employee ID: ")
dept= input("Department: ")

emp = Employee(name, id, dept)
emp.disp_id()
```

```python
#Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def disp(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.disp()


```

```python
#Problem 4: Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def tag(self):
        print(f"\nProduct: {self.name}")
        print(f"Price: ₹{self.price}")

name = input("Product Name: ")
price = float(input("Price: "))

product = Product(name, price)
product.tag()
```

```python
#Problem 5: Movie Rating Display System

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def disp(self):
        print(f"\nMovie: {self.name}")
        print(f"Rating: {self.rating} / 5")

name = input("Movie Name: ")
rating = float(input("Rating: "))

movie = Movie(name, rating)
movie.disp()

```

```python
#Problem 6: Delivery Address Manager

class Customer:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def details(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")


name = input("Customer Name: ")
address = input("Address: ")

customer = Customer(name, address)
customer.details()

```
